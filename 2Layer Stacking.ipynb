```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import *
from sklearn.ensemble import *
from sklearn.datasets import make_classification
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression , SGDRegressor
from sklearn.svm import SVR

from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv('data/energy-train-l.csv').values
X_train3, y_train3 = data[:, [0]].copy(), data[:, [1]].copy()

x_normalizer = StandardScaler()
y_normalizer = StandardScaler()

X_train3 = x_normalizer.fit_transform(X_train3)
y_train3 = y_normalizer.fit_transform(y_train3)

data = pd.read_csv('data/energy-test-l.csv').values
X_test3, y_test3 = data[:, [0]].copy(), data[:, [1]].copy()
X_test3 = x_normalizer.transform(X_test3)
y_test3 = y_normalizer.transform(y_test3)

print(X_train3.shape, X_test3.shape, y_train3.shape, y_test3.shape)

plt.scatter(X_train3, y_train3, c='red', marker='x')
plt.scatter(X_test3, y_test3, c='blue', marker='x')
plt.legend(['Train', 'Test']);
```

```python
model1 = StackingRegressor(estimators=[('knn', KNeighborsRegressor(n_neighbors=45)),
                                       ('rf', RandomForestRegressor(n_estimators=45, random_state=42 , ccp_alpha=0.1)),
                                       ('svm', SVR())],
                           final_estimator=LinearRegression()
                          )
model1.fit(X_train3, y_train3.ravel())



model2 = StackingRegressor(estimators=[('knn', KNeighborsRegressor(n_neighbors=45)),
                                       ('rf', RandomForestRegressor(n_estimators=45, random_state=42 , ccp_alpha=0.1)),
                                       ('svm', SVR())],
                           final_estimator=DecisionTreeRegressor(max_depth=3)
                          )
model2.fit(X_train3, y_train3.ravel())
                           
                           
                           
model3 = StackingRegressor(estimators=[('knn', KNeighborsRegressor(n_neighbors=45)),
                                       ('rf', RandomForestRegressor(n_estimators=45, random_state=42 , ccp_alpha=0.1)),
                                       ('svm', SVR())],
                           final_estimator=GradientBoostingRegressor(n_estimators=20)
                          )
model3.fit(X_train3, y_train3.ravel())
                           
                     
                           
final_model = StackingRegressor(estimators=[('model1', model1),
                                       ('model2', model2),
                                       ('model3', model3)],
                           final_estimator=SGDRegressor()
                               )
final_model.fit(X_train3, y_train3.ravel())                        
```

```python
print(f'Score for model1 is {model1.score(X_test3, y_test3)}')
print(f'Score for model2 is {model2.score(X_test3, y_test3)}')
print(f'Score for model3 is {model3.score(X_test3, y_test3)}')                             
                                
print(f'Score for final_model is {final_model.score(X_test3, y_test3)}')
```

```python
fig , axes = plt.subplots(2 , 2 , figsize=(12 , 8))
ax_flat = axes.flatten()

models = [model1 , model2 , model3 , final_model]
names = ['model 1' , 'model 2' , 'model 3' , 'final model']

xp = np.linspace(X_train3.min(), X_train3.max(), 100)[:, None]

for i , model in enumerate(models):
    
    ax_flat[i].scatter(X_train3, y_train3, c='red' , ec='k')
    ax_flat[i].scatter(X_test3, y_test3, c='blue' , ec='k')


    yp = model.predict(xp)
    ax_flat[i].plot(xp, yp, c='gold', linewidth=3)

    ax_flat[i].set_title(f'Result For {names[i]}' , fontfamily='Serif' , fontweight='bold')
    ax_flat[i].legend(['Train', 'Test']);
    
    
```
